```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2015 = pd.read_excel('WITS-Partner_2015.xls', sheet_name="Partner")
export_2016 = pd.read_excel('WITS-Partner_2016.xls', sheet_name="Partner")
```

```python
export_2016.head()
```

```python
export = pd.concat([export_2015, export_2016])
```

```python
export.sample(10)
```

```python
export.pivot_table(index=['Year'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
exclude = ['Europe & Central Asia', 'East Asia & Pacific', 'Unspecified', ' World']
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
.sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```
